```python
import plotly.express as px
df = px.data.iris()
fig = px.histogram(df, x = "sepal_length" , y = "petal_width")
fig.show()
```

```python
import plotly.express as px
df = px.data.iris()
fig = px.scatter(df, x= "species", y = "petal_width")
fig.show()
```

```python
import plotly.express as px
df = px.data.iris()
fig = px.scatter(df, x= "species", y = "petal_width", size = "petal_length", color = "species")
fig.show()
```
